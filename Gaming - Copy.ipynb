```python
#convert float to int
data.NA_players=(data.NA_players.round(0).astype(int))
data.EU_players=(data.EU_players.round(0).astype(int))
data.Other_players=(data.Other_players.round(0).astype(int))
data.JP_players=(data.JP_players.round(0).astype(int))

```

```python
plt.figure(figsize=(10,7))

#find percentage
data_perc1 = data.NA_players.divide(data.sum(axis=1), axis=0)
data_perc2 = data.JP_players.divide(data.sum(axis=1), axis=0)
data_perc3 = data.EU_players.divide(data.sum(axis=1), axis=0)
data_perc4 = data.Other_players.divide(data.sum(axis=1), axis=0)

data_perc= (data_perc1+data_perc2+data_perc3+data_perc4)/4
#plotgraph
plt.stackplot(data.Year_of_Release,data.NA_players,data.JP_players,data.EU_players,data.Other_players,
              edgecolor='black',
              labels=['North America','Japan','Europe','Others'])

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
